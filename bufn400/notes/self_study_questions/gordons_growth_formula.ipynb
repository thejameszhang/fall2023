# Problem 1
Suppose that dividends are paid continuously, as a flow, and $\bar{r}$
is a continuously compounded expected return. Verify intuitively (or with math) the expected result that Gordon's growth formula becomes
$$P(t_n) = \frac{D(t_n)}{\bar{r} - \bar{g}}$$
(This is yet another reason why continuously-compounded rates are convenient for theoretical modeling.)

**Solution:**

Recall that the denominator is not dependent on $n$ in the summation. Thus, it is like a constant and we can pull it out of the summation. This is due to one of the assumptions with Gordon's Growth Formula, which is the expected return of the risky asset and the growth rate of the dividends are constant over time. Mathematically, this makes sense because the closer $\bar{r}$ and $\bar{g}$ are, the greater the price is, and this makes intuitively because that means the growth rate of diviends and the expected return of the asset are both large.

# Problem 2
2. To replicate all of the dividends, the number of share bought should be equal to one. Verify this by verifying that
$$\sum_{n=1}^\infty x_n = 1$$
(If this equation were not true, stocks would have a positive or negative bubble component, where a bubble is defined as a component of value which does not represent the present value of cash flows.)

**Solution:**

Recall that 

$$x_n = \frac{\bar{r} - \bar{g}* \Delta t}{((1 + \bar{g} * \Delta t))} (\frac{1 + \bar{g} * \Delta t}{1 + \bar{r} + \Delta t})^n$$

and so by subsitution we get that we have to show

$$\sum_{n=1}^\infty \frac{\bar{r} - \bar{g}* \Delta t}{(1 + \bar{g} * \Delta t)} (\frac{1 + \bar{g} * \Delta t}{1 + \bar{r} + \Delta t})^n = 1$$

Observe that the left fraction can be pulled out of the summation, so we get

$$\frac{\bar{r} - \bar{g}* \Delta t}{(1 + \bar{g} * \Delta t)} \sum_{n=1}^\infty (\frac{1 + \bar{g} * \Delta t}{1 + \bar{r} + \Delta t})^n = 1$$

$$ \sum_{n=1}^\infty (\frac{1 + \bar{g} * \Delta t}{1 + \bar{r} + \Delta t})^n = \frac{1 + \bar{g} * \Delta t}{(\bar{r} - \bar{g}* \Delta t)}$$

# Problem 3

Explain what goes wrong if $\bar{r} - \bar{g}$
 is negative. (Obviously, the price cannot be negative, so something else is happening.)

**Solution:**

If $\bar{r} - \bar{g}$ goes negative, then yes obviously recall that the price cannot be negative. Further recall that Gordon's Growth Formula can be expressed as 

$$PV[D_n * \Delta t, n * \Delta t | \mathcal{H}(t_0)] = D_0 * \Delta t * (\frac{1 + \bar{g} * \Delta t}{1 + \bar{r} * \Delta t})^n$$

In this form, we can see that if $\bar{r} - \bar{g} \implies \bar{g} > \bar{r}$, which means the constant expected dividend growth rate is greater the the constant expected return of the risky asset. Thus, the numerator in the fraction grows faster than the denominator, and so Gordon's Growth Formula predicts that the price of the risky asset goes to $\infty$, which obviously is also not the case.



# Problem 4
When modeling, is it better to work with the cap rate (dividend yield) $\kappa$
 or its reciprocal (price-dividend ratio) $1/\kappa$
, when working with different stocks with different values of $\kappa$?

**Solution:**

If you are interested in comparing the income generated by different stocks relative to their price, then the cap rate (dividend yield) $\kappa$ would be more appropriate. This is because the dividend yield directly measures the return on investment from dividends.

On the other hand, if you are more interested in the valuation of the stock relative to its income (dividends), then the price-dividend ratio $1/\kappa$ would be more suitable. This ratio is often used in valuation models and can provide insights into whether a stock is overvalued or undervalued based on its income generation.
